# Problema 1
 
Desarrolle una función la cual reciba dos parámetros: una imagen y un entero llamado color. La función debe devolver una imagen que tenga activos los canales de color según los siguientes puntos:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("p1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show
```

```python
def procesar_imagen(img, color):
    
    nueva_imagen = np.zeros_like(img)

    if color == 1:
        # Azul
        nueva_imagen[:, :, 0] = img[:, :, 0]
    elif color == 2:
        # Verde
        nueva_imagen[:, :, 1] = img[:, :, 1]
    elif color == 3:
        # Rojo
        nueva_imagen[:, :, 2] = img[:, :, 2]
    elif color == 10:
        # Rojo y Verde
        nueva_imagen[:, :, 1] = img[:, :, 1]  # Verde
        nueva_imagen[:, :, 2] = img[:, :, 2]  # Rojo
    elif color == 20:
        # Verde y Azul
        nueva_imagen[:, :, 0] = img[:, :, 0]  # Azul
        nueva_imagen[:, :, 1] = img[:, :, 1]  # Verde
    elif color == 30:
        # Azul y Rojo
        nueva_imagen[:, :, 0] = img[:, :, 0]  # Azul
        nueva_imagen[:, :, 2] = img[:, :, 2]  # Rojo
    else:
        raise ValueError("Valor no válido. Debe ser un color con valor 1, 2, 3, 10, 20, o 30.")

    return nueva_imagen
```

```python
# Procesar la imagen con el parámetro "color" con valor = 1
resultado = procesar_imagen(img, 1)
# Mostrar la imagen resultante
plt.imshow(cv2.cvtColor(resultado, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def crear_imagenes_grises(imagen_rgb_path):
    # Leer la imagen a color
    imagen_rgb = cv2.imread("p3_original.jpg")
    
    # Verifica que la imagen fue cargada correctamente
    if imagen_rgb is None:
        raise ValueError("La imagen no se pudo cargar. Verifique la ruta.")

    # Convertir la imagen a formato RGB (OpenCV usa BGR por defecto)
    imagen_rgb = cv2.cvtColor(imagen_rgb, cv2.COLOR_BGR2RGB)

    # Separar los canales de color
    canal_rojo = imagen_rgb[:, :, 0]
    canal_verde = imagen_rgb[:, :, 1]
    canal_azul = imagen_rgb[:, :, 2]

    # Crear las imágenes en escala de grises a partir de cada canal
    imagen_gris_rojo = np.stack((canal_rojo,) * 3, axis=-1)
    imagen_gris_verde = np.stack((canal_verde,) * 3, axis=-1)
    imagen_gris_azul = np.stack((canal_azul,) * 3, axis=-1)

    # Mostrar las imágenes en escala de grises
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].imshow(imagen_gris_rojo)
    axs[0].set_title('Escala de Grises - Canal Rojo')
    axs[0].axis('off')
    
    axs[1].imshow(imagen_gris_verde)
    axs[1].set_title('Escala de Grises - Canal Verde')
    axs[1].axis('off')
    
    axs[2].imshow(imagen_gris_azul)
    axs[2].set_title('Escala de Grises - Canal Azul')
    axs[2].axis('off')

    plt.show()

    # Guardar las imágenes resultantes
    cv2.imwrite('problema3_imagen_gris_rojo.jpg', cv2.cvtColor(imagen_gris_rojo, cv2.COLOR_RGB2BGR))
    cv2.imwrite('problema3_imagen_gris_verde.jpg', cv2.cvtColor(imagen_gris_verde, cv2.COLOR_RGB2BGR))
    cv2.imwrite('problema3_imagen_gris_azul.jpg', cv2.cvtColor(imagen_gris_azul, cv2.COLOR_RGB2BGR))

# Ruta a la imagen a color
ruta_imagen_rgb = 'problema3_imagen.jpg'

# Crear y mostrar las imágenes en escala de grises
crear_imagenes_grises(ruta_imagen_rgb)
```

```python
def mostrar_histogramas(imagen_path):
    # Leer la imagen a color
    imagen = cv2.imread("p3_original.jpg")
    
    # Verificar que la imagen fue cargada correctamente
    if imagen is None:
        raise ValueError("La imagen no se pudo cargar. Verifique la ruta.")
    
    # Convertir la imagen a RGB (OpenCV usa BGR por defecto)
    imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

    # Separar los canales de color
    canal_rojo = imagen_rgb[:, :, 0]
    canal_verde = imagen_rgb[:, :, 1]
    canal_azul = imagen_rgb[:, :, 2]

    # Calcular la imagen en escala de grises usando promedio aritmético
    imagen_gris = (canal_rojo + canal_verde + canal_azul) / 3
    imagen_gris = imagen_gris.astype(np.uint8)  # Convertir a tipo uint8

    # Calcular histogramas y medias
    def calcular_histograma_y_media(canal):
        histograma, bins = np.histogram(canal.flatten(), bins=256, range=[0,256])
        media = np.mean(canal)
        return histograma, media

    # Histograma y media para cada canal de color
    histograma_rojo, media_rojo = calcular_histograma_y_media(canal_rojo)
    histograma_verde, media_verde = calcular_histograma_y_media(canal_verde)
    histograma_azul, media_azul = calcular_histograma_y_media(canal_azul)
    
    # Histograma y media para la imagen en escala de grises
    histograma_gris, media_gris = calcular_histograma_y_media(imagen_gris)

    # Mostrar los histogramas
    fig, axs = plt.subplots(2, 2, figsize=(15, 10))
    
    axs[0, 0].plot(histograma_rojo, color='red')
    axs[0, 0].axvline(media_rojo, color='red', linestyle='--')
    axs[0, 0].set_title('Histograma Canal Rojo')
    axs[0, 0].set_xlim([0, 255])

    axs[0, 1].plot(histograma_verde, color='green')
    axs[0, 1].axvline(media_verde, color='green', linestyle='--')
    axs[0, 1].set_title('Histograma Canal Verde')
    axs[0, 1].set_xlim([0, 255])

    axs[1, 0].plot(histograma_azul, color='blue')
    axs[1, 0].axvline(media_azul, color='blue', linestyle='--')
    axs[1, 0].set_title('Histograma Canal Azul')
    axs[1, 0].set_xlim([0, 255])

    axs[1, 1].plot(histograma_gris, color='gray')
    axs[1, 1].axvline(media_gris, color='black', linestyle='--')
    axs[1, 1].set_title('Histograma Escala de Grises')
    axs[1, 1].set_xlim([0, 255])

    plt.tight_layout()
    plt.show()

# Ruta a la imagen a color
ruta_imagen = 'p4_imagen.jpg'

# Mostrar los histogramas
mostrar_histogramas(ruta_imagen)
```

##### El enfoque de escala de grises ponderado consiste en convertir una imagen a escala de grises utilizando una combinación ponderada de los canales de color (rojo, verde y azul). En lugar de simplemente promediar los valores de estos canales, se asignan diferentes pesos a cada uno para reflejar mejor cómo los humanos perciben la intensidad luminosa de cada color.

### Los pesos más comunes que se utilizan son los siguientes:

##### Rojo (R): 0.299
##### Verde (G): 0.587
##### Azul (B): 0.114
##### Estos pesos están basados en la sensibilidad del ojo humano a los colores, donde el verde contribuye más a la luminosidad percibida, seguido por el rojo, y finalmente el azul.

<div style="text-align: center;">
    <h1>Espacio de Color HSV</h1>
</div>

<p style="text-align: justify;">
El espacio de color <strong>HSV</strong> (Hue, Saturation, Value) es un modelo de color que representa los colores de una manera más alineada con la percepción humana en comparación con el modelo RGB (Rojo, Verde, Azul). Este espacio es especialmente útil en aplicaciones donde se necesita manipular el color de manera intuitiva, como en la edición de imágenes y la visión por computadora.
</p>

<h2 style="text-align: center;">Componentes del Espacio HSV</h2>

<ul>
    <li><strong>Hue (Tono):</strong> Representa el tipo de color, o su "matiz". Se mide en grados, desde 0° a 360°, donde:
        <ul>
            <li>0° corresponde al rojo.</li>
            <li>120° corresponde al verde.</li>
            <li>240° corresponde al azul.</li>
        </ul>
    </li>
    <li><strong>Saturation (Saturación):</strong> Indica la pureza del color o cuán "vibrante" es. Va desde 0% (gris) hasta 100% (color completamente puro).</li>
    <li><strong>Value (Valor):</strong> Representa la luminosidad del color, o cuán brillante es. Va desde 0% (negro) hasta 100% (el color en su máxima luminosidad).</li>
</ul>

<h2 style="text-align: center;">Mapeo de Colores al Espacio HSV</h2>

<p style="text-align: justify;">
Para convertir un color desde el espacio RGB al espacio HSV, se sigue un proceso matemático que normaliza los valores RGB entre 0 y 1 y luego calcula el tono, la saturación y el valor en función de las diferencias entre los valores máximos y mínimos de los componentes RGB.
</p>

<ul>
    <li><strong>Hue:</strong> Determinado por la diferencia máxima y mínima entre los valores RGB y su posición relativa.</li>
    <li><strong>Saturation:</strong> Calculado como la diferencia entre el valor máximo y mínimo normalizado por el valor máximo.</li>
    <li><strong>Value:</strong> El valor máximo de los componentes RGB.</li>
</ul>
